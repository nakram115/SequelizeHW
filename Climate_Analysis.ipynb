```python
# SQL Alchemy
from sqlalchemy import create_engine

#PyMySQL 
#import pymysql
#pymysql.install_as_MySQLdb()

from sqlalchemy import create_engine, func
#from sqlalchemy.ext.declarative import declarative_base


#from sqlalchemy import Column, Integer, String, Float, Text, Numeric, ForeignKey

from sqlalchemy.orm import Session
from sqlalchemy.ext.automap import automap_base

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

#from flask import Flask, jsonify
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
# Reflect database into ORM class
Base = automap_base()
Base.prepare(engine, reflect=True)
```

```python
Base.classes.keys()
```

```python
# Measurements class
Measurements = Base.classes.measurement
```

```python
Stations = Base.classes.station
```

```python
# Start a session to query the database
session = Session(engine)
```

```python
#only have these dates as 12 months from data
start_date= "2016-08-23"
end_date= "2017-08-23"

session.query(Measurements.date, Measurements.prcp).filter(Measurements.date<=end_date).filter(Measurements.date>=start_date).all()
```
